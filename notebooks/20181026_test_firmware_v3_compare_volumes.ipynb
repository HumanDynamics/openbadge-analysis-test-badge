Testing volumes in the two versions of the badges. Using these badges:
Old firmware:
* C0:93:A3:F4:48:B4 IYK0OK3TBN
* CA:AB:C4:D0:E4:94 XUOL644SQH

New:
* F6:C8:4C:08:A8:40 ERV129QAGP
* F2:29:0D:98:93:62 A974F3I63H



```python
from __future__ import absolute_import, division, print_function

import os
import sys
import datetime


import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("ggplot")

# import data analysis modules
import openbadge_analysis as ob
import openbadge_analysis.core
```

```python
log_file_path = '../data/raw/20181026_test_firmware_v3_compare_volumes/audio_archive.txt'
```

```python
df_meeting = ob.core.sample2data(log_file_path,log_version = '2.0')
```

```python
df_meeting
```

```python
#Old firmware:
#* C0:93:A3:F4:48:B4 IYK0OK3TBN
#* CA:AB:C4:D0:E4:94 XUOL644SQH
#
#New:
#* F6:C8:4C:08:A8:40 ERV129QAGP
#* F2:29:0D:98:93:62 A974F3I63H
#o1 = df_meeting.loc['IYK0OK3TBN']
#o2 = df_meeting.loc['ERV129QAGP']
o1 = df_meeting.loc['XUOL644SQH']
o2 = df_meeting.loc['A974F3I63H']
time_slice = slice('2018-10-26 16:21:42', '2018-10-26 16:22:00')
o1 = o1.loc[time_slice, :]
o2 = o2.loc[time_slice, :]
```

```python
o1['ticks']=range(0,len(o1.index.values))
o2['ticks']=range(0,len(o2.index.values))
```

```python
o1.reset_index().plot(x='datetime', y='signal')
o2.reset_index().plot(x='datetime', y='signal')
```

```python
o1.reset_index().plot.scatter(x='ticks', y='signal',figsize=(15,10))
o2.reset_index().plot.scatter(x='ticks', y='signal',figsize=(15,10))
```

```python
o1t = o1.iloc[100:190]
o2t = o2.iloc[100:190]
```

```python
o1t.reset_index().plot.scatter(x='ticks', y='signal',figsize=(10,7))
o2t.reset_index().plot.scatter(x='ticks', y='signal',figsize=(10,7))
```
